### Plotting global accuracy vs communication rounds for K = 5

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,201,12)

path = './Results/'

temp = np.load(path+'history_acc1.npy')
hist_acc1 = [temp[i]*100 for i in range(0,201,12)]
temp = np.load(path+'history_acc2.npy')
hist_acc2 = [temp[i]*100 for i in range(0,201,12)]
temp = np.load(path+'history_acc3.npy')
hist_acc3 = [temp[i]*100 for i in range(0,201,12)]
temp = np.load(path+'history_acc4.npy')
hist_acc4 = [temp[i]*100 for i in range(0,201,12)]

plt.plot(x, hist_acc1, label='B=10, E=1')
plt.plot(x, hist_acc2, label='B=10, E=5')
plt.plot(x, hist_acc3, label='B=20, E=1')
plt.plot(x, hist_acc4, label='B=20, E=5')
# Naming the x-axis, y-axis and the whole graph
plt.xlabel("No. of communication rounds")
plt.ylabel("Global Accuracy (%)")
plt.title("CNN Classifier and K = 5")
  
# Adding legend, which helps us recognize the curve according to it's color
plt.legend()
  
# To load the display window
plt.show()
```
