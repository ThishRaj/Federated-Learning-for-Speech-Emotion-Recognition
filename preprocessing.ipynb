```python
import librosa
import numpy as np
import os
```

```python
path = './Datasets/RawData/'
mylist = os.listdir(path)
print("Path has %d files" %(len(mylist)))
```

```python
def extract_mfccs(filename):
    audio, Fs = librosa.load(filename, duration=2.25, sr=22050, offset=0.5)
    features = librosa.feature.mfcc(y=audio, sr=22050, n_mfcc = 40)
    
    return features
```

```python
labels = []
feeling_nos = ['01','02','04','05']
feeling_labels = [0,1,2,3]
feeling_names = ['neutral', 'happy', 'sad', 'angry']
feeling_map = dict(zip(feeling_nos, feeling_labels))
count = [0,0,0,0]
print(feeling_map)
flag = 1
for full_filename in mylist:
    
    filename = full_filename[0:20]
    # print(filename)

    filename=filename.split('-')
    # print(filename)
    emotion = filename[2]
    if emotion in feeling_map:
        labels.append(feeling_map[emotion]) 
        if flag == 1:
            i = 0
            temp = np.mean(extract_mfccs(path+full_filename), axis=0)
            mfccs = np.ndarray((644,temp.shape[1]))
            flag = 0
        
        print(i,full_filename, temp.shape)
        temp = np.mean(extract_mfccs(path+full_filename), axis=0)
        mfccs[i,:] = temp.T
        i=i+1

print(labels)
```

```python
from sklearn.preprocessing import OneHotEncoder
from keras.utils import np_utils
y = np_utils.to_categorical(labels)
print(y.shape)
print(y)
```

```python
import random
print(y.shape[0])
np.random.seed(1)
shuffled_indices = np.arange(y.shape[0])
np.random.shuffle(shuffled_indices)
print(shuffled_indices)
```

```python
train_indices = shuffled_indices[np.arange(int(y.shape[0]*0.8))]
test_indices = np.array(list(set(shuffled_indices) - set(train_indices)))
print(test_indices.shape)
print(train_indices.shape)
```

```python
y_train = y[train_indices]
x_train = mfccs[train_indices]
y_test = y[test_indices]
x_test = mfccs[test_indices]

print(y_train.shape)
print(y_test.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
np.save('./Datasets/x_train.npy',x_train)
np.save('./Datasets/x_test.npy',x_test)
np.save('./Datasets/y_train.npy',y_train)
np.save('./Datasets/y_test.npy',y_test)
```
